## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
day_df["total_users"] = day_df["casual"] + day_df["registered"]

day_df["dteday"] = pd.to_datetime(day_df["dteday"])

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.scatter(day_df["temp"], day_df["total_users"], alpha=0.5, color="blue")
plt.title("Temperature vs Bike Rentals")
plt.xlabel("Temperature")
plt.ylabel("Total Rentals")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.scatter(day_df["hum"], day_df["total_users"], alpha=0.5, color="red")
plt.title("Humidity vs Bike Rentals")
plt.xlabel("Humidity")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.scatter(day_df["windspeed"], day_df["total_users"], alpha=0.5, color="green")
plt.title("Windspeed vs Bike Rentals")
plt.xlabel("Windspeed")
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
day_df["month"] = day_df["dteday"].dt.month

monthly_trends = day_df.groupby("month")["total_users"].sum().reset_index()

month_labels = {
    1: "Januari",
    2: "Februari",
    3: "Maret",
    4: "April",
    5: "Mei",
    6: "Juni",
    7: "Juli",
    8: "Agustus",
    9: "September",
    10: "Oktober",
    11: "November",
    12: "Desember"
}
monthly_trends["month"] = monthly_trends["month"].map(month_labels)

plt.figure(figsize=(10, 5))
plt.plot(monthly_trends["month"], monthly_trends["total_users"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Monthly Bike Rental Trends", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Rentals")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
