## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
day_df["total_users"] = day_df["casual"] + day_df["registered"]


day_df["day_type"] = day_df["workingday"].map({0: "Akhir Pekan/Libur", 1: "Hari Kerja"})


day_type_trends = day_df.groupby("day_type")["total_users"].mean().reset_index()


plt.figure(figsize=(10, 5))
plt.bar(day_type_trends["day_type"], day_type_trends["total_users"], color=["#FF7F50", "#4682B4"])
plt.title("Rata-rata Peminjaman Sepeda: Hari Kerja vs Akhir Pekan", fontsize=14)
plt.xlabel("Tipe Hari")
plt.ylabel("Rata-rata Jumlah Peminjam")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()


plt.figure(figsize=(10, 5))
day_type_users = day_df.groupby("day_type")[["casual", "registered"]].mean().reset_index()


plt.bar(day_type_users["day_type"], day_type_users["casual"], color="#FF7F50", label="Casual")

plt.bar(day_type_users["day_type"], day_type_users["registered"], color="#4682B4",
        bottom=day_type_users["casual"], label="Registered")

plt.title("Komposisi Peminjam Sepeda: Hari Kerja vs Akhir Pekan", fontsize=14)
plt.xlabel("Tipe Hari")
plt.ylabel("Rata-rata Jumlah Peminjam")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()


plt.figure(figsize=(10, 5))

total_by_type = day_df.groupby("day_type")[["casual", "registered"]].sum()

percentage = total_by_type.div(total_by_type.sum(axis=1), axis=0) * 100
percentage = percentage.reset_index()


plt.subplot(1, 2, 1)
plt.pie(percentage.loc[percentage["day_type"] == "Akhir Pekan/Libur", ["casual", "registered"]].values[0],
        labels=["Casual", "Registered"], autopct='%1.1f%%', colors=["#FF7F50", "#4682B4"], startangle=90)
plt.title("Komposisi Peminjam\nAkhir Pekan/Libur")

plt.subplot(1, 2, 2)
plt.pie(percentage.loc[percentage["day_type"] == "Hari Kerja", ["casual", "registered"]].values[0],
        labels=["Casual", "Registered"], autopct='%1.1f%%', colors=["#FF7F50", "#4682B4"], startangle=90)
plt.title("Komposisi Peminjam\nHari Kerja")

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
day_df["month"] = day_df["dteday"].dt.month

monthly_trends = day_df.groupby("month")["total_users"].sum().reset_index()

month_labels = {
    1: "Januari",
    2: "Februari",
    3: "Maret",
    4: "April",
    5: "Mei",
    6: "Juni",
    7: "Juli",
    8: "Agustus",
    9: "September",
    10: "Oktober",
    11: "November",
    12: "Desember"
}
monthly_trends["month"] = monthly_trends["month"].map(month_labels)

plt.figure(figsize=(10, 5))
plt.plot(monthly_trends["month"], monthly_trends["total_users"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Monthly Bike Rental Trends", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Rentals")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Bagaimana pengaruh cuaca terhadap jumlah peminjam sepeda

1.   Perbedaan Volume Peminjaman

  *   Terdapat perbedaan jumlah peminjam sepeda antara hari kerja dan akhir pekan/hari libu
  *  Rata-rata jumlah peminjam pada hari kerja lebih tinggi dibandingkan dengan akhir pekan
  * Pola ini konsisten dan menunjukkan bahwa sepeda lebih banyak digunakan untuk aktivitas rutin harian seperti perjalanan ke tempat kerja



2. Perbedaan Komposisi Pengguna

  *   Pengguna terdaftar (registered) mendominasi peminjaman pada hari kerja, menunjukkan pola penggunaan yang lebih teratur
  *  Pengguna kasual (casual) memiliki proporsi yang lebih tinggi pada akhir pekan/hari libur
  
  *  Persentase pengguna casual pada akhir pekan hampir dua kali lipat dibandingkan pada hari kerja








- Conclution pertanyaan 2 : Penyewaan sepeda meningkat di musim panas dan menurun di musim dingin.


1.   Pola Musiman yang Jelas


  *    Terdapat pola musiman yang sangat jelas dalam penggunaan sepeda sepanjang tahun
  *   Jumlah peminjaman sepeda mencapai puncaknya pada bulan-bulan musim panas (Juni-September)
  * Jumlah peminjaman terendah terjadi pada bulan-bulan musim dingin (Desember-Februari)

2.  Tren Tahunan
  * Terjadi peningkatan penggunaan yang signifikan dari musim semi hingga musim panas
  * Puncak peminjaman terjadi pada bulan Agustus dan September
